## Preprocessing

```python
# Step: Preprocessing
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Make a copy
df_processed = df.copy()

# 1. Encode categorical variables
label_encoders = {}
for col in df_processed.columns:
    if df_processed[col].dtype == "object":
        le = LabelEncoder()
        df_processed[col] = le.fit_transform(df_processed[col])
        label_encoders[col] = le  # save encoder for later (e.g. decoding predictions)

print("Categorical features encoded.")

# 2. Remove irrelevant features (example: Unnamed index columns, IDs if present)
to_drop = [c for c in df_processed.columns if c.lower().startswith("unnamed")]
df_processed = df_processed.drop(columns=to_drop, errors="ignore")

print(f"Irrelevant columns removed: {to_drop}")

# 3. Shuffle the dataset
df_processed = df_processed.sample(frac=1, random_state=42).reset_index(drop=True)

print("Dataset shuffled.")
print("\nProcessed dataset shape:", df_processed.shape)
df_processed.head()
```

##  Train-Test Split and Model Training

```python

# Step: Train/Test Split + Model Training & Evaluation

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.neural_network import MLPClassifier

# Features (X) and Target (y)
X = df_processed.drop("Class", axis=1)
y = df_processed["Class"]

# Split into train/test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, shuffle=True
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)

# Define models
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Perceptron": Perceptron(max_iter=1000, random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=5000, random_state=42),
    "Neural Network (MLP)": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
}

# Train and evaluate each model
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"\n{name} Results:")
    print("Accuracy:", round(acc, 4))
    print(classification_report(y_test, y_pred))
```

## Custom Prediction Function

```python

 
# Step: Custom Prediction Function
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# --- Load data ---
file_path = r"C:\Users\User\Downloads\AI-Data.csv"
df = pd.read_csv(file_path)

# --- Basic preprocessing ---
# Drop missing values (you can customize this)
df = df.dropna()

# Separate features and target
X = df.drop("Class", axis=1)
y = df["Class"]

# One-hot encode categorical features
X_encoded = pd.get_dummies(X)

# Save the feature columns for later use with custom input
feature_columns = X_encoded.columns

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Split into train/test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, shuffle=True
)

# --- Train models ---
models = {
    "LogisticRegression": LogisticRegression(max_iter=500),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC()
}

for name, model in models.items():
    model.fit(X_train, y_train)

# --- Evaluate models ---
print("Model Accuracies:")
for name, model in models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name}: {acc:.4f}")

# --- Prediction function for custom input ---
def predict_custom_input(input_data, models, feature_columns, scaler):
    """
    input_data: dict of feature_name -> value (must match original dataset columns, except 'Class')
    models: dictionary of trained models
    feature_columns: columns used during training (after encoding)
    scaler: fitted StandardScaler object
    """
    sample_df = pd.DataFrame([input_data])
    
    # One-hot encode the sample input's categorical features
    sample_encoded = pd.get_dummies(sample_df)
    
    # Align columns with training data
    sample_encoded = sample_encoded.reindex(columns=feature_columns, fill_value=0)
    
    # Scale the input
    sample_scaled = scaler.transform(sample_encoded)
    
    # Predict with all models
    predictions = {}
    for name, model in models.items():
        pred = model.predict(sample_scaled)[0]
        predictions[name] = pred
    
    return predictions

# --- Example custom input ---
custom_input = {
    "gender": "M",
    "NationalITy": "KW",
    "PlaceofBirth": "KuwaIT",
    "StageID": "lowerlevel",
    "GradeID": "G-04",
    "SectionID": "A",
    "Topic": "IT",
    "Semester": "F",
    "Relation": "Father",
    "raisedhands": 20,
    "VisITedResources": 15,
    "AnnouncementsView": 5,
    "Discussion": 10,
    "ParentAnsweringSurvey": "Yes",
    "ParentschoolSatisfaction": "Good",
    "StudentAbsenceDays": "Under-7"
}

# Run prediction on custom input
preds = predict_custom_input(custom_input, models, feature_columns, scaler)
print("\nPredictions for custom input:")
for model_name, prediction in preds.items():
    print(f"{model_name}: {prediction}")
```
